### **1. Finns det ett samband mellan antalet storkar och antalet f√∂dslar?**  
Nej, det finns inget orsakssamband mellan storkar och barnaf√∂dsel. Detta √§r ett exempel p√• **korrelation**, men inte **kausalitet**.  

- **Korrelation** betyder att tv√• saker h√§nder samtidigt.  
- **Kausalitet** betyder att en sak orsakar en annan.  

I detta fall √§r det troligt att b√•de antalet storkar och f√∂dslar √§r vanligare p√• landsbygden, vilket g√∂r att de verkar ha ett samband ‚Äì men storkar orsakar inte f√∂dslar.

### **2: Vad √§r korrelation?**  

Korrelation betyder att tv√• saker f√∂r√§ndras tillsammans, men det betyder inte att den ena orsakar den andra.  

#### **Exempel 1: Glassf√∂rs√§ljning och drunkningsolyckor**  
- N√§r glassf√∂rs√§ljningen √∂kar, √∂kar ocks√• antalet drunkningsolyckor.  
- Detta betyder inte att glass orsakar drunkning.  
- B√•da √∂kar eftersom det √§r varmt p√• sommaren (**en tredje faktor p√•verkar b√•da**).  

#### **Exempel 2: L√§ngd och skostorlek**  
- Personer med st√∂rre skostorlek √§r ofta l√§ngre.  
- H√§r finns en **korrelation**, men det betyder inte att stora f√∂tter g√∂r dig l√§ngre ‚Äì b√•da p√•verkas av din kropps tillv√§xt.

#### **Typer av korrelation**  
- **Positiv korrelation**: N√§r en variabel √∂kar, √∂kar √§ven den andra.  
  - Exempel: Ju mer du tr√§nar, desto b√§ttre kondition f√•r du.  
- **Negativ korrelation**: N√§r en variabel √∂kar, minskar den andra.  
  - Exempel: Ju mer tid du spenderar p√• sociala medier, desto mindre tid har du f√∂r studier.  

**Viktigt:** Korrelation betyder inte **kausalitet** (orsakssamband).  


### **3. √Ñr medelv√§rde b√§ttre √§n median?**  
Det beror p√• situationen!  

- **Medelv√§rde** (genomsnitt) √§r k√§nsligt f√∂r extremv√§rden.  
- **Median** (mittenv√§rdet) p√•verkas inte av extremv√§rden.  

Exempel:  
Om vi har dessa l√∂ner: **25 000, 26 000, 27 000, 28 000, 1 000 000 kr**  
- **Medelv√§rde:** (25 000 + 26 000 + 27 000 + 28 000 + 1 000 000) √∑ 5 = **221 200 kr**  
- **Median:** **27 000 kr** (mittenv√§rdet)  

H√§r ger medianen en mer r√§ttvis bild av l√∂nerna.

### **7. Vad betyder att nyf√∂ddas vikt √§r normalf√∂rdelad?**  

Normalf√∂rdelning betyder att de flesta v√§rden ligger n√§ra medelv√§rdet, och att v√§ldigt h√∂ga eller l√•ga v√§rden √§r ovanliga.
